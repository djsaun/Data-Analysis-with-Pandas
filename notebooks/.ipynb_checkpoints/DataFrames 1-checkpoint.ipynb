## Broadcasting Operations

```python
nba["Age"].add(5)
nba["Age"] + 5 ## adds 5
```

```python
nba["Salary"].sub(5000)
nba["Salary"] - 5000 ## subtracts 5000
```

```python
nba["Weight"].mul(2)
nba["Weight"] * 2 ## multiplies by 2
```

```python
## save new series as new column within DataFrame 
nba["Weight in Kilograms"] = nba["Weight"] * 0.453592
nba.head(3)
```

```python
nba["Salary"].div(1000000)
nba["Salary in Millions"] = nba["Salary"] / 1000000 ## divides by 1,000,000
nba.head(3)
```
